```python
#Discount Usage Frequency (DUF)
def calculate_duf(df):
    """
    Calculates the Discount Usage Frequency (DUF) for each customer.

    Args:
        df: A pandas DataFrame containing the following columns:
            - order_id
            - customer_id
            - discount_applied (boolean or amount)

    Returns:
        A pandas Series with customer_id as index and DUF as values.
    """

    # Filter orders with discounts
    discounted_orders = df[df['discount_applied'] > 0]
    # print(discounted_orders)
    # Count orders per customer
    order_counts = df.groupby('customer_id')['order_id'].count()
    # print(order_counts)
    # Count discounted orders per customer
    discounted_order_counts = discounted_orders.groupby('customer_id')['order_id'].count()
    # print(discounted_order_counts)
    # Fill missing values with 0 (customers with no discounted orders)
    discounted_order_counts = discounted_order_counts.reindex(order_counts.index).fillna(0)
    # print(discounted_order_counts)
    # Calculate DUF
    duf = discounted_order_counts / order_counts
    # print(duf)
    return duf
k=calculate_duf(df)
k
```

```python
import pandas as pd

def calculate_dscaf(df):
    """
    Calculates the Discount Sensitive Cart Abandonment Frequency (DSCAF) for each customer.

    Args:
        df (pd.DataFrame): A DataFrame containing the following columns:
            - cart_id: Unique identifier for each cart.
            - customer_id: Unique identifier for each customer.
            - cart_status: Indicates whether the cart was 'abandoned' or 'completed'.
            - discount_viewed: Boolean indicating whether a discount was viewed in the cart.

    Returns:
        pd.Series: A Series with customer_id as index and DSCAF values as values. NaN values are replaced with 0.
    """

    # Handle missing values
    df = df.fillna({
        'cart_status': 'completed',  # Assume missing cart_status means completed (not abandoned)
        'discount_viewed': False    # Assume missing discount_viewed means no discount was viewed
    })
    # Ensure that data types are correct
    df['discount_viewed'] = df['discount_viewed'].astype(bool)

    # Filter relevant carts (abandoned and discount viewed)
    abandoned_discount_viewed = df[
        (df['cart_status'] == 'abandoned') & (df['discount_viewed'] == True)
    ]

    # Filter all carts where a discount was viewed
    all_discount_viewed = df[df['discount_viewed'] == True]

    # Count occurrences per customer
    abandoned_counts = abandoned_discount_viewed['customer_id'].value_counts()
    total_counts = all_discount_viewed['customer_id'].value_counts()

    # Calculate DSCAF and handle division by zero (replace with 0)
    dscaf_series = abandoned_counts.divide(total_counts, fill_value=0)

    return dscaf_series
print(calculate_dscaf(df))
```

```python
import pandas as pd

def calculate_discount_proportion(df: pd.DataFrame) -> pd.Series:
    """
    Calculates the Discount Proportion for each customer.

    This function determines the proportion of discounted items a customer
    has purchased out of their total items purchased.

    Args:
        df (pd.DataFrame): A Pandas DataFrame containing the following columns:
            - order_id: Unique identifier for each order
            - item_id: Unique identifier for each item within an order
            - discount_applied: Indicates whether a discount was applied to an item
                (True, False, or a numerical discount amount)

    Returns:
        pd.Series: A Pandas Series containing the Discount Proportion for each
                   unique `customer_id` in the DataFrame.
    """
   
    # Handle missing values
    df["discount_applied"] = df["discount_applied"].fillna(0)

    # Mark items as discounted if `discount_applied` is not 0 
    df["is_discounted"] = df["discount_applied"].apply(lambda x: True if x > 0 else False)

    # Calculate discount proportion per customer
    discount_proportion = (
        df.groupby('customer_id')['is_discounted']
        .agg(lambda x: x.mean())
        .fillna(0)  # Customers with no items get a 0% discount proportion
    )

    return discount_proportion
print(calculate_discount_proportion(df))
```

```python
import pandas as pd

def calculate_pfds(df: pd.DataFrame) -> pd.Series:
    """
    Calculates the Purchase Frequency During Sales (PFDS) for each customer.

    This function determines the proportion of a customer's purchases that were
    made during sale periods, indicating their tendency to buy when items are
    discounted.

    Args:
        df (pd.DataFrame): A Pandas DataFrame containing the following columns:
            - order_id: Unique identifier for each order
            - customer_id: Unique identifier for each customer
            - order_date: Datetime value indicating the date of purchase
            - sale_period: Boolean (True/False) or datetime range 
                           indicating whether the purchase was made during a sale

    Returns:
        pd.Series: A Pandas Series containing the PFDS for each unique
                   `customer_id` in the DataFrame, ranging from 0 to 1.
    """

    # Handle missing values in `sale_period`
    # Assuming a purchase is NOT during a sale if `sale_period` is missing
    df['sale_period'] = df['sale_period'].fillna(False)

    # If `sale_period` is a datetime range, convert to boolean
    if pd.api.types.is_datetime64_any_dtype(df['sale_period']):
        df['sale_period'] = df['sale_period'].apply(
            lambda x: True
            if x.left <= df['order_date'] <= x.right
            else False
        )
    
    # Calculate total purchases per customer
    total_purchases = df.groupby('customer_id')['order_id'].nunique()

    # Filter to purchases made during sales
    sale_purchases = df[df['sale_period']].groupby('customer_id')['order_id'].nunique()

    # Calculate PFDS and fill missing values with 0 (customers with no sale purchases)
    pfds = sale_purchases.div(total_purchases, fill_value=0)

    return pfds
print(calculate_pfds(df))
```
