1. Run the cell to generate classification for any customer ID (1-100)

```python
from discount_hunter_classifier import run
#read csv file into dataframe
import pandas as pd
import numpy as np
df=pd.read_csv(r".\synthetic_ecommerce_data.csv")
run(df[df['customer_id']==18])

```

2. Run this cell to generate classification for all customers and save as a csv

```python
output_list = []
unique_customer_ids = df['customer_id'].unique()

for customer_id in unique_customer_ids:
    customer_orders = df[df['customer_id'] == customer_id].to_dict('records')
    result = run(customer_orders)
    result['customer_id'] = customer_id  # Add customer_id to the result for reference
    output_list.append(result)

# Create a DataFrame with the output
output_df = pd.DataFrame(output_list, columns=[
    'customer_id', 'is_discount_hunter', 'score', 'duf', 
    'discount_proportion', 'pfds', 'dscaf'
])

# Save the DataFrame to a CSV file
output_df.to_csv(r"output.csv", index=False)
```

3. Run this cell for finding the average discount hunter score - or any other analysis for setting the threshold value.


```python
#find average discount hunter for all customers
avg = 0
for i in df['customer_id'].unique():
    output=run(df[df['customer_id']==i])
    print(output)
    score=output['score']
    avg += score
print("Average Discount Hunter Score: ",avg/len(df['customer_id'].unique()))
```
